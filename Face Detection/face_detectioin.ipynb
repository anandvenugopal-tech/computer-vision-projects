Import libraries

```python
import cv2
import mediapipe as mp
import os
```

Initialize mediapipe face detection

```python
my_face = mp.solutions.face_detection
my_draw = mp.solutions.drawing_utils
```

```python
face_detection = my_face.FaceDetection()
```

Open webcam & detect face

```python
cap = cv2.VideoCapture(0)

while True:
    success, img = cap.read()
    if not success:
        break
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    results = face_detection.process(img_rgb)

    if results.detections:
        for det in results.detections:
            my_draw.draw_detection(img, det)
    
    cv2.imshow("Face Detection",img)

    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

Face Detection on a image

```python
img2 = cv2.imread('family_image.jpg')
img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
results = face_detection.process(img2_rgb)

if results.detections:
    for det in results.detections:
        my_draw.draw_detection(img2, det)
    
cv2.imshow('Face Detection', img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
