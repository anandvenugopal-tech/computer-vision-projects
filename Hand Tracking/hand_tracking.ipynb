Import libraries

```python
import mediapipe as mp
import cv2
```

Initialize mediapipe hand tracking

```python
mp_hands = mp.solutions.hands
mp_draw = mp.solutions.drawing_utils
```

```python
hands = mp_hands.Hands(
    max_num_hands = 2
)
```

Open webcam & Detect hands

```python
cap = cv2.VideoCapture(0)

while True:
    success, img = cap.read()
    if not success:
        break
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    results = hands.process(img_rgb)

    if results.multi_hand_landmarks:    
        for hand_landmarks in results.multi_hand_landmarks:
            mp_draw.draw_landmarks(img, hand_landmarks, mp_hands.HAND_CONNECTIONS)
    
    cv2.imshow('Hand Tracking', img)

    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()

```

Hand detectin on an image

```python
img2 = cv2.imread('hand.jpg.jpeg')
img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

results = hands.process(img2_rgb)

if results.multi_hand_landmarks:
    for hand_landmarks in results.multi_hand_landmarks:
        mp_draw.draw_landmarks(img2, hand_landmarks, mp_hands.HAND_CONNECTIONS)

cv2.imshow('Hand Detection', img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
